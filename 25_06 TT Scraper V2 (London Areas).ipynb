### Need to pull in the list of towns
Let's see how many foodbanks this gets us to
List can be found here: https://www.paulstenning.com/uk-towns-and-counties-list/

```python
towns_list = pd.read_csv('London Areas.csv')
towns_list.describe()
```

```python
url_string = 'https://www.trusselltrust.org/get-help/find-a-foodbank/'

results = []

for i in range(540):
    try:
        url = url_string+str(towns_list.Location[i]).replace(" ", "")
        source = request.urlopen(url).read().decode()
        soup = bs4.BeautifulSoup(source, 'html')
        
        title = [title.text.strip() for title in soup.findAll('h1', {'class': 'heading--leader green'})]
        #get the email address:
        email = [email.text.strip() for email in soup.findAll('a', href=re.compile('^mailto'))]
        #get the phone number:
        phone = [phone.text.strip() for phone in soup.findAll('a', href=re.compile('^tel:'))]
        
        results.append({'title' : str(title), 'email' : str(email), 'phone' : str(phone)})
        
        print('{} FOUND!'.format(url))
        x+=1
        
        #don't overflow website
        sleep(5)
        
    except urllib.error.HTTPError:
        print('{} not found'.format(url))
        x+=1
        sleep(3)
        continue
```

```python
#results
df = pd.DataFrame.from_dict(results)

df.iloc[:, 0] = df.iloc[:, 0].str[0].astype(str)
df.iloc[:, 1] = df.iloc[:, 1].str[0].astype(str)
df.iloc[:, 2] = df.iloc[:, 2].str[0].astype(str)

#df.to_csv('foodbanks_London.csv', mode='w')
```
