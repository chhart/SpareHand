### Need to pull in the list of towns
Let's see how many foodbanks this gets us to
List can be found here: https://www.paulstenning.com/uk-towns-and-counties-list/

```python
foodbanks_list = pd.read_csv('All_Foodbanks.csv')
foodbanks_list.head()
```

```python
if 'results' in locals():
    pass
else:
    results = []

if 'x' in locals():
    pass
else:
    x=0

for i in range(x,429):
    try:
        url = foodbanks_list['URL'][i]
        source = request.urlopen(url)
        soup = bs4.BeautifulSoup(source, 'html')
        
        try:
            title = soup.find('h1', {'class': 'heading--leader green'}).text.strip()
        except:
            pass
        
        #get the email address:
        try:
            email = soup.find('a', href=re.compile('^mailto')).text.strip()
        except:
            pass
        
        #get the phone number:
        try:
            phone =  soup.find('a', href=re.compile('^tel:')).text.strip()
        except:
            pass
        
        results.append({'title' : str(title), 'email' : str(email), 'phone' : str(phone)})
        
        print('{} FOUND!'.format(url))
        x+=1
        
        #don't overflow website
        sleep(3)
        
    except urllib.error.HTTPError:
        print('{} not found'.format(url))
        x+=1
        sleep(3)
        continue
```

```python
#results
df = pd.DataFrame.from_dict(results)

# write this all to a csv
df.to_csv('foodbanks_FINAL.csv', mode='w')
```
